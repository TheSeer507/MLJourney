To Visualize our model predictions lets create a function plot decision boundary.
* Take in trained model, features and labels
* create a meshgrid of the different x values
* make predictions across the meshgrid
* plot the prediction as a la line between the zones

```python
import numpy as np


def plot_decision_boundary(model, X, y):
    """ 
    Plots the decision boundary created by a model predicting on X.
    """
    #Define the axis boundaries of the plot
    x_min, x_max = X[:,0].min() - 0.1, X[:,0].max() + 0.1
    y_min, y_max = X[:,1].min() - 0.1, X[:,1].max() + 0.1

    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100),
                                    np.linspace(y_min, y_max, 100))
    #Create X values
    x_in = np.c_[xx.ravel(), yy.ravel()] #stack 2d array together

    #Make prediction
    y_pred = model.predict(x_in)

    #check for multiclass
    if len(y_pred[0]) > 1:
        print("Doing Multiclass Classification")
        y_pred = np.argmax(y_pred, axis=1).reshape(xx.shape)
    else:
        print("Doing Binary Classification")
        y_pred = np.round(y_pred).reshape(xx.shape)
    #plot the decision boundary
    plt.contourf(xx, yy, y_pred, cmap=plt.cm.RdYlBu, alpha=0.7)
    plt.scatter(X[:,0], X[:,1], c=y, s=40, cmap=plt.cm.RdYlBu)
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
```

```python
#Check the predictions our model is making
plot_decision_boundary(model=model_3, X=X, y=y)
```

```python
#Let's see if our model can be use for regression problem

tf.random.set_seed(42)

#Create some regression data
X_regression = tf.range(0,1000, 5)
y_regression = tf.range(100, 1100, 5) #y = x + 100

X_regression = tf.reshape(X_regression, (-1, 1)) # Reshape to 2 dimensions
y_regression = tf.reshape(y_regression, (-1, 1)) # Reshape to 2 dimensions
```

```python
#Split our regression Data into training and test

X_reg_train = X_regression[:150]
X_reg_test = X_regression[150:]
y_reg_train = y_regression[:150]
y_reg_test = y_regression[150:]

#fit our model
#Model_3 was compiled for Binnary classification now we are working with regression model needs to be changed
```

```python
#set up random seed

tf.random.set_seed(42)

model_4 = tf.keras.Sequential([
    tf.keras.layers.Dense(100),
    tf.keras.layers.Dense(10),
    tf.keras.layers.Dense(1)
])

#Compile model
model_4.compile(loss=tf.keras.losses.mae,
                optimizer=tf.keras.optimizers.Adam(),
                metrics=["mae"])

model_4.fit(X_reg_train, y_reg_train, epochs=100)
```

```python
#Make Predictions
y_reg_preds = model_4.predict(X_reg_test)

#plot the model
plt.figure(figsize=(10,7))

plt.scatter(X_reg_train, y_reg_train, c='b', label='Training Data')
plt.scatter(X_reg_test, y_reg_test, c='g', label='Test Data')
plt.scatter(X_reg_test, y_reg_preds, c='r', label='Predictions')
plt.legend();
```

```python
#Create a toy tensor

A = tf.cast(tf.range(-10, 10), tf.float32)
A
```

```python
#Visualize our toy tensor
plt.plot(A);
```

```python
#Replicate sigmoid sigmoid(x) = 1 / 1(1+exp(-x))
def sigmoid(x):
    return 1 / (1+tf.exp(-x))

#use the sigmoid function
sigmoid(A)
```

```python
#Plot our toy tensor to use sigmoid
plt.plot(sigmoid(A));
```

```python
#Lets recreate Relu Function
def relu(x):
    return tf.maximum(0, x)

#pass toy tensor to our custom relu function
relu(A)
```

```python
#plot ReLU tensor
plt.plot(relu(A));
```
