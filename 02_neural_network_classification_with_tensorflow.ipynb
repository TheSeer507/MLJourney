To Visualize our model predictions lets create a function plot decision boundary.
* Take in trained model, features and labels
* create a meshgrid of the different x values
* make predictions across the meshgrid
* plot the prediction as a la line between the zones

```python
import numpy as np


def plot_decision_boundary(model, X, y):
    """ 
    Plots the decision boundary created by a model predicting on X.
    """
    #Define the axis boundaries of the plot
    x_min, x_max = X[:,0].min() - 0.1, X[:,0].max() + 0.1
    y_min, y_max = X[:,1].min() - 0.1, X[:,1].max() + 0.1

    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100),
                                    np.linspace(y_min, y_max, 100))
    #Create X values
    x_in = np.c_[xx.ravel(), yy.ravel()] #stack 2d array together

    #Make prediction
    y_pred = model.predict(x_in)

    #check for multiclass
    if len(y_pred[0]) > 1:
        print("Doing Multiclass Classification")
        y_pred = np.argmax(y_pred, axis=1).reshape(xx.shape)
    else:
        print("Doing Binary Classification")
        y_pred = np.round(y_pred).reshape(xx.shape)
    #plot the decision boundary
    plt.contourf(xx, yy, y_pred, cmap=plt.cm.RdYlBu, alpha=0.7)
    plt.scatter(X[:,0], X[:,1], c=y, s=40, cmap=plt.cm.RdYlBu)
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
```

```python
#Check the predictions our model is making
plot_decision_boundary(model=model_3, X=X, y=y)
```

### Finding the best learning rate

To find the ideal learning rate (the learning rate where the loss decreases the most during training) we're going to use the following steps:
* A learning rate callback - you can think of a callback as an extra piece of functionality, you can add to your *while* its training.
* Setup a series of learning rates to test (we'll use 10 different learning rates here).
* Another model (identical to the one we created back in model_1) but this time with a very high learning rate, this model will be used to find the ideal learning rate.
* A Modified loss curve plot - this will plot the loss versus the learning rate (instead of steps).

```python
#set random seed
tf.random.set_seed(42)

#create the model
model_9 = tf.keras.Sequential([
    tf.keras.layers.Dense(4, activation="relu"),
    tf.keras.layers.Dense(4, activation="relu"),
    tf.keras.layers.Dense(1, activation="sigmoid")
])

#compile the model
model_9.compile(loss=tf.keras.losses.BinaryCrossentropy(),
                optimizer=tf.keras.optimizers.Adam(),
                metrics=["accuracy"])

#Create a learning rate callback
lr_scheduler = tf.keras.callbacks.LearningRateScheduler(lambda epoch: 1e-4 * 10**(epoch/20))

#fit the model 
history_9 = model_9.fit(X_train, y_train, epochs=100, callbacks=[lr_scheduler])
```

```python
#Check the history
pd.DataFrame(history_9.history).plot(figsize=(10,7), xlabel="epochs");
```

```python
#Plot the learning rate versus the loss
lrs = 1e-4 * (10**(tf.range(100)/20))
plt.figure(figsize=(10,7))
plt.semilogx(lrs, history_9.history["loss"])
plt.xlabel("Learning Rate")
plt.ylabel("Loss")
plt.title("Learning rate vs Loss");
```

Rule of thumb: Ideal Learning rate is where the loss decreases the most, the lowest point of the loss curve.

```python
# Ideal learning rate is around 0.02
#Example of other learning rate values
10**0, 10**-1, 10**-2, 10**-3, 10**-4
```
